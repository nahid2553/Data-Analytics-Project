```python
import pandas as pd
import numpy as np
import matplotlib as mlb
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stock_data= pd.read_csv('gdrive/My Drive/colabfile/Stock_Market_Data.csv')
stock_data.head()
```

```python
stock_data["Date"]= pd.to_datetime(stock_data["Date"],dayfirst=True)
stock_data.dtypes
```

```python
stock_data["Date"].max()
```

**Creating List of Spacific Company**

```python
specific_company=['UNILEVERCL','BEXIMCO', 'RECKITTBN','BATBC', 'SAIHAMCOT']

## Filter out desired companies data and put it to another data frame
specific_data=stock_data[stock_data['Name'].isin(specific_company)]

##creating Line Graph for each companies
plt.figure(figsize=[12,6])
for company in specific_company:
  company_data= specific_data[specific_data['Name']==company]
  plt.plot(company_data['Date'],company_data['Close'],label=company )

## Adding labels and titles
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Stock Price Trend For Specific Company')
plt.legend()
plt.grid()

## Improving Readablity
plt.xticks(rotation=45)
plt.show()
```

```python
specific_company='RECKITTBEN'

specific_data=stock_data[stock_data['Name']==specific_company]
specific_data['7_Day_Rolling_Avg']=specific_data['Close'].rolling(window=7).mean()
plt.figure(figsize=[12,6])

plt.plot(specific_data['Date'],specific_data['Close'], label=f'{specific_company} Closing Price',color='Blue' )

plt.plot(specific_data['Date'],specific_data['7_Day_Rolling_Avg'], label=f'{specific_company} 7_Day_Rolling_Avg of Closing Price',color='Red' )

plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title(f'{specific_company} 7 Day Rolling Avarage of closing price')
plt.grid()
plt.legend()

plt.xticks(rotation=45)
plt.show()
```
